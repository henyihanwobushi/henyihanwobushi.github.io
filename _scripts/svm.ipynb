```python
def get_split_by_entropy_gain(data):
    sorted_data = sorted(data, key=lambda x: x[0], reverse=True)
    return [gini(sorted_data, i / 100.) for i in range(1, 100)]

def information_gain(sorted_data, split_point):
    cnts = [[0, 0], [0, 0]]
    for d in sorted_data:
        cnts[0 if d[0] < split_point else 1][0 if d[1] == 0 else 1] += 1
    


def gini(sorted_data, split_point):

    p1 = 1.0 * cnts[0][1] / (cnts[0][0] + cnts[0][1])
    p2 = 1.0 * cnts[1][1] / (cnts[1][0] + cnts[1][1])
    return 1 - (p1**2 + p2**2)


# print(df.values.tolist())
sd = get_split_by_entropy_gain(df.values.tolist())
print(sd)
```
